```python
# train_all / test_all / val_all 파일들에 각 mbti 개수 출력
import pandas as pd
import matplotlib.pyplot as plt
import os

# === 1) 파일 경로 지정 ===
base_dir = "/Users/seojunha/PycharmProjects/MBTI_Prediction/Data/split"
files = ["train.csv", "val.csv", "test.csv"]

# === 2) 모든 CSV 불러와서 합치기 ===
dfs = []
for f in files:
    path = os.path.join(base_dir, f)
    df = pd.read_csv(path)
    df["split"] = f.replace(".csv", "").capitalize()  # 구분용 컬럼 추가
    dfs.append(df)

data = pd.concat(dfs, ignore_index=True)
print(f"Loaded total {len(data):,} rows from all splits.\n")

# === 3) MBTI 정리 ===
data["mbti"] = data["mbti"].astype(str).str.upper().str.strip()

# === 4) MBTI별 개수 및 비율 계산 ===
counts = data["mbti"].value_counts().sort_index()
percentages = (counts / len(data) * 100).round(2).astype(str) + "%"

# === 5) 결과 출력 ===
print("=== MBTI Counts ===")
print(counts)
print("\nTotal samples:", len(data))
print("\n=== Percentage Distribution ===")
print(percentages)

# === 6) 시각화 ===
plt.figure(figsize=(10, 5))
plt.bar(counts.index, counts.values, color="skyblue", edgecolor="black")
plt.title("MBTI Distribution (Train + Val + Test Combined)", fontsize=14)
plt.xlabel("MBTI Type")
plt.ylabel("Number of Posts")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```
