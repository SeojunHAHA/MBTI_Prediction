```python
import pandas as pd
import matplotlib.pyplot as plt

# === 1) Load your dataset ===
data = pd.read_csv("/Users/seojunha/PycharmProjects/MBTI_Prediction/Data/processed/reddit_post_clean.csv")

# === 2) Clean and normalize MBTI column ===
data["mbti"] = data["mbti"].astype(str).str.upper().str.strip()

# === 3) Count each MBTI type ===
counts = data["mbti"].value_counts().sort_index()

# === 4) Print the counts ===
print("=== MBTI Counts ===")
print(counts)
print("\nTotal samples:", len(data))
print("\n=== Percentage Distribution ===")
print((counts / len(data) * 100).round(2).astype(str) + "%")

# === 5) Plot a bar chart ===
plt.figure(figsize=(10,5))
plt.bar(counts.index, counts.values, color="skyblue", edgecolor="black")
plt.title("MBTI Distribution in Reddit Dataset", fontsize=14)
plt.xlabel("MBTI Type")
plt.ylabel("Number of Posts")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```
